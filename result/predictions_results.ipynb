```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("predictions_max.csv")

y_true = df["tm_max_price"]
y_pred = df["prediction"]

plt.figure(figsize=(8, 6))
plt.scatter(y_true, y_pred, alpha=0.6, color='blue', edgecolors='w', s=60)
plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--', linewidth=2)
plt.xlabel("Actual Max Price")
plt.ylabel("Predicted Max Price")
plt.title("Predicted vs Actual Max Ticket Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter out outliers
df_plot = df[(df["tm_max_price"] <= 1000) & (df["prediction"] <= 1000)]

y_true = df_plot["tm_max_price"]
y_pred = df_plot["prediction"]

plt.figure(figsize=(8, 6))
plt.scatter(y_true, y_pred, alpha=0.6, color='blue', edgecolors='w', s=60)
plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--', linewidth=2)
plt.xlabel("Actual Max Price")
plt.ylabel("Predicted Max Price")
plt.title("Predicted vs Actual Max Ticket Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```
